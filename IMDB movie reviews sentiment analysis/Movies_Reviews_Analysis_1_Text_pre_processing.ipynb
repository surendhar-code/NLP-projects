**Text Pre-Processing**
1. Removing punctuation marks.
2. Removing stopwords - here I don't remove stopwords because in stock news each and every word is impoertant.
3. Converting the words into vectors using Bag of words and TF-IDF methods.

```
#Storing each reviews in a sentence list

sentences=[]
for i in range(0,len(data['review'])):
  sentences.append(data['review'][i])
```

```
from bs4 import BeautifulSoup 
import re
import nltk
nltk.download('punkt')
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
final_sent=[]
for i in range(0,50000):
  review_text=BeautifulSoup(sentences[i]).get_text() #to remove html tags
  review_text=re.sub('[^a-zA-Z]',' ',review_text)    #to remove punctuation marks
  review_text=review_text.lower().split()            #lowering the sentence and word tokenization
  words=[word for word in review_text if word not in set(stopwords.words('english'))] #removing stopwords
  review_text=' '.join(words) #joining the word after pre-processing
  final_sent.append(review_text)
```

```
#Printing the final_sentence after pre-processing
final_sent[48000]
```

```
#converting list of sentences into dataframe.
processed_data=pd.DataFrame(final_sent)
```
